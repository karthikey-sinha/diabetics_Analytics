```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('diabetese.tsv',sep='\t')
df.head()
```

```python
#Let's Find the Avg Age of diabetic Patients
mean_age = df['AGE'].mean()
print(f"Avg Age of diabetic Patients : {mean_age}")
```

```python
#variance
var_age = df['AGE'].var()
print(f"Variance of Age is : {var_age}")
```

```python
#Standard Deviation
std_age = df['AGE'].std()
print(f"Standard Deviation of Age is :{std_age}")
```

```python
plt.figure(figsize=(2,6))
plt.boxplot(df['AGE'] ,showmeans=True)
plt.grid(color='gray', linestyle = 'dotted')
plt.xlabel('Age')
plt.ylabel("Years")
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.boxplot(df['BMI'],vert=False,showmeans=True)
plt.grid( color='gray', linestyle = 'dotted')
plt.xlabel('Value')
plt.ylabel("BMI")
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.boxplot(df['BP'],vert = False, showmeans=True)
plt.grid(color='gray', linestyle= 'dotted')
plt.xlabel('Value')
plt.ylabel('Blood Pressure')
plt.show()
```

```python
# What age of people are Mostly Affected by Diabetes?
# Creating Histogram
df['AGE'].hist(bins=12)
plt.title('Age Distribution of Diabetic Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# What age of people are Mostly Affected by Diabetes?
# Creating Histogram
df['BMI'].hist(bins=12)
plt.title('BMI Distribution of Diabetic Patients')
plt.xlabel('BMI')
plt.ylabel('Number of Patients')
plt.show()
```

```python
hours_studies = [2,3,1,4,2,5,4,6,3,5]
exam_scores = [70,80,60,85,75,90,88,92,80,95]

plt.scatter(hours_studies, exam_scores, color='Red', marker='o')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Scatter Plot of Hours Studied vs Exam Scores')
plt.show()
```

```python
plt.scatter(df['BMI'], df['Y'] , color='Blue')

plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Scatter Plot : BMI vs Disease Progression')
plt.show()

```

```python
plt.scatter(df['S5'], df['Y'] , color='Blue')

plt.xlabel('S5 : Blood Type')
plt.ylabel('Disease Progression')
plt.title('Scatter Plot : S5 vs Disease Progression')
plt.show()

```
